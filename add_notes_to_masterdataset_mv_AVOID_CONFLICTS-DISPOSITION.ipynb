```python
import pandas as pd

TIME_COL      = "outtime"
NOTE_TIME_COL = "charttime"
TEXT_COL      = "text"
OUT_COL       = "discharge_notes"
OUT_ID_COL    = "discharge_note_id"

NORMALIZE_TEXT = True

master = master_dataset.copy()
notes  = discharge.copy()

master[TIME_COL] = pd.to_datetime(master[TIME_COL], errors="coerce")
notes[NOTE_TIME_COL] = pd.to_datetime(notes[NOTE_TIME_COL], errors="coerce")

master["subject_id"] = pd.to_numeric(master["subject_id"], errors="coerce").astype("Int64")
notes["subject_id"]  = pd.to_numeric(notes["subject_id"],  errors="coerce").astype("Int64")

master = master.dropna(subset=["subject_id", TIME_COL]).copy()
notes  = notes.dropna(subset=["subject_id", NOTE_TIME_COL, TEXT_COL]).copy()

master["subject_id"] = master["subject_id"].astype("int64")
notes["subject_id"]  = notes["subject_id"].astype("int64")

master["_row_id"] = master.index.astype("int64")

max_outtime = master[TIME_COL].max()
notes = notes[notes[NOTE_TIME_COL] <= max_outtime].copy()

if NORMALIZE_TEXT:
    notes["_text_key"] = (
        notes[TEXT_COL].astype(str)
        .str.replace(r"\s+", " ", regex=True)
        .str.strip()
    )
else:
    notes["_text_key"] = notes[TEXT_COL].astype(str)

master_sorted = master.sort_values([TIME_COL, "subject_id"], kind="mergesort").reset_index(drop=True)
notes_sorted  = notes.sort_values([NOTE_TIME_COL, "subject_id"], kind="mergesort").reset_index(drop=True)

cols_from_notes = ["subject_id", NOTE_TIME_COL, TEXT_COL, "_text_key"]
if "note_id" in notes_sorted.columns:
    cols_from_notes.insert(3, "note_id")

cand = pd.merge_asof(
    master_sorted,
    notes_sorted[cols_from_notes],
    by="subject_id",
    left_on=TIME_COL,
    right_on=NOTE_TIME_COL,
    direction="backward",
    allow_exact_matches=True,
)

matched = cand.dropna(subset=["_text_key"]).copy()
matched["_gap_sec"] = (matched[TIME_COL] - matched[NOTE_TIME_COL]).dt.total_seconds()

final_assignments = (
    matched.sort_values(["_text_key", "_gap_sec"], kind="mergesort")
           .drop_duplicates("_text_key", keep="first")
)

keep_cols = ["_row_id", TEXT_COL]
rename_map = {TEXT_COL: OUT_COL}

if "note_id" in final_assignments.columns:
    keep_cols.insert(1, "note_id")
    rename_map["note_id"] = OUT_ID_COL

mapping = final_assignments[keep_cols].rename(columns=rename_map)

master_out = master_dataset.copy()
master_out["_row_id"] = master_out.index.astype("int64")

master_out = master_out.merge(mapping, on="_row_id", how="left").drop(columns=["_row_id"])
master_dataset = master_out
```

```python
import pandas as pd

def chunk_generator(df, chunk_size=10000):
    for start in range(0, len(df), chunk_size):
        yield df.iloc[start:start + chunk_size].copy()

def safe_combine(d_note, r_note):
    if pd.isna(d_note) and pd.isna(r_note):
        return pd.NA
    elif pd.isna(d_note):
        return r_note
    elif pd.isna(r_note):
        return d_note
    else:
        return f"{d_note} {r_note}"

def process_and_save(master_dataset, output_path, chunk_size=10000):
    with open(output_path, 'w', encoding='utf-8') as f_out:
        header_written = False

        for chunk in chunk_generator(master_dataset, chunk_size):
            # Process chunk
            chunk['notes'] = [
                safe_combine(d, r)
                for d, r in zip(chunk['discharge_notes'], chunk['radiology_notes'])
            ]

            # Write to CSV incrementally
            chunk.to_csv(f_out, index=False, header=not header_written)
            header_written = True

process_and_save(master_dataset, 'master_dataset_notes_DISPOSITION.csv', chunk_size=10000)

```
