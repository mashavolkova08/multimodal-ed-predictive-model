```python
df_ecg = pd.read_csv("master_dataset_ecg_DISPOSITION.csv")
print(len(df_ecg))
df_ecg
```

```python
import pandas as pd
import ast

df = df_ecg

df["ecg_paths"] = df["ecg_paths"].apply(ast.literal_eval)

df["last_ecg_path"] = df["ecg_paths"].apply(lambda x: x[-1] if isinstance(x, list) and len(x) > 0 else None)

g = df.groupby(["subject_id", "last_ecg_path"])

conflicts = g.filter(
    lambda x: (
        x["outcome_critical"].nunique() > 1
        or x["outcome_hospitalization"].nunique() > 1
    )
)

conflicts
```

```python
df_notes = pd.read_csv("master_dataset_notes_cleaned_DISPOSITION.csv")
print(len(df_notes))
df_notes
```

```python
df_notes.columns.to_list()
```

```python
conflict_hosp_notes = (
    df_notes.groupby("notes")["outcome_hospitalization"]
    .nunique()
    .loc[lambda x: x > 1]
    .index
)

# find notes with conflicting critical outcomes
conflict_crit_notes = (
    df_notes.groupby("notes")["outcome_critical"]
    .nunique()
    .loc[lambda x: x > 1]
    .index
)

# combine both sets
conflict_notes = set(conflict_hosp_notes).union(set(conflict_crit_notes))

print("Total conflicting note texts:", len(conflict_notes))

# remove them
df_notes_clean = df_notes[~df_notes["notes"].isin(conflict_notes)].copy()

print("Rows before:", len(df_notes))
print("Rows after:", len(df_notes_clean))
```

```python
structured_train["split"] = "train"
structured_test["split"] = "test"

structured_full = pd.concat([structured_train, structured_test], ignore_index=True)
```

```python
structured_full = structured_full.merge(
    df_notes[["stay_id", "notes"]],
    on="stay_id",
    how="left"
)

structured_full = structured_full.merge(
    df_ecg[["stay_id", "ecg_paths"]],
    on="stay_id",
    how="left"
)
```

```python
structured_full
```

```python
structured_full = structured_full[
    structured_full["notes"].notna() & structured_full["ecg_paths"].notna()
]
```

```python
# shuffle rows
structured_full = structured_full.sample(frac=1, random_state=3180).reset_index(drop=True)

n = len(structured_full)

train_end = int(0.8 * n)
val_end = int(0.9 * n)

train_df = structured_full.iloc[:train_end]
val_df   = structured_full.iloc[train_end:val_end]
test_df  = structured_full.iloc[val_end:]
```

```python
train_df.to_csv("mv_train_DISPOSITION.csv", index=False)
val_df.to_csv("mv_val_DISPOSITION.csv", index=False)
test_df.to_csv("mv_test_DISPOSITION.csv", index=False)
```
