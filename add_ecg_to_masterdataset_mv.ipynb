```python
import pandas as pd
import numpy as np

# ONE ECG PER VISIT, PRE-VISIT ONLY, ECG USED AT MOST ONCE

# For each subject, process visits in chronological order.
# Assign the latest unused ECG with ecg_time < visit intime.

master_dataset = master_dataset.copy()
df_ecg = df_ecg.copy()

master_dataset["intime"]  = pd.to_datetime(master_dataset["intime"])
master_dataset["outtime"] = pd.to_datetime(master_dataset["outtime"])
df_ecg["ecg_time"]        = pd.to_datetime(df_ecg["ecg_time"])

master_dataset = master_dataset.reset_index(drop=False).rename(columns={"index": "visit_key"})

visits = master_dataset[["visit_key", "subject_id", "intime", "outtime"]].copy()
ecgs   = df_ecg[["subject_id", "ecg_time", "path"]].copy()

ecgs = ecgs.reset_index(drop=False).rename(columns={"index": "ecg_key"})

assigned_rows = []

for sid, visits_sub in visits.groupby("subject_id", sort=False):
    ecgs_sub = ecgs[ecgs["subject_id"] == sid].copy()

    if visits_sub.empty:
        continue

    visits_sub = visits_sub.sort_values("intime").copy()
    ecgs_sub   = ecgs_sub.sort_values("ecg_time").copy()

    used_ecg_keys = set()

    for _, vrow in visits_sub.iterrows():
        intime = vrow["intime"]

        eligible = ecgs_sub[
            (ecgs_sub["ecg_time"] < intime) &
            (~ecgs_sub["ecg_key"].isin(used_ecg_keys))
        ]

        if eligible.empty:
            continue

        chosen = eligible.iloc[-1]

        used_ecg_keys.add(chosen["ecg_key"])

        assigned_rows.append({
            "visit_key": vrow["visit_key"],
            "subject_id": sid,
            "ecg_key": chosen["ecg_key"],
            "ecg_time": chosen["ecg_time"],
            "path": chosen["path"],
            "distance_to_intime": intime - chosen["ecg_time"]
        })

assigned = pd.DataFrame(assigned_rows)

if not assigned.empty:
    assigned = assigned.sort_values(["visit_key", "distance_to_intime"], ascending=[True, False])

    grouped_ecg = (
        assigned.groupby("visit_key")["path"]
        .apply(lambda x: list(x.astype(str)))
        .reset_index()
        .rename(columns={"path": "ecg_paths"})
    )
else:
    grouped_ecg = pd.DataFrame(columns=["visit_key", "ecg_paths"])

master_dataset = (
    master_dataset
    .drop(columns=["ecg_paths", "last_ecg_path"], errors="ignore")
    .merge(grouped_ecg, on="visit_key", how="left")
)

master_dataset["last_ecg_path"] = master_dataset["ecg_paths"].apply(
    lambda x: x[-1] if isinstance(x, list) and len(x) > 0 else None
)

print("Total visits:", len(master_dataset))
print("Visits with assigned ECG:", master_dataset["last_ecg_path"].notna().sum())
print("Visits without assigned ECG:", master_dataset["last_ecg_path"].isna().sum())

if not assigned.empty:
    n_dup = int((assigned.groupby("ecg_key")["visit_key"].nunique() > 1).sum())
    print("ECGs assigned to >1 visit:", n_dup)

    # confirm all are pre-visit
    n_bad = int((assigned.merge(visits[["visit_key", "intime"]], on="visit_key")["ecg_time"] >=
                 assigned.merge(visits[["visit_key", "intime"]], on="visit_key")["intime"]).sum())
    print("Assigned ECGs with ecg_time >= intime:", n_bad)

    display(
        assigned.sort_values(["subject_id", "ecg_time"]).head(10)
    )


```
